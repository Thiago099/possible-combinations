```python
max_len = 6
possible_chars = ['1', '9']
for j in range(1,max_len + 1):
    for i in range(pow(len(possible_chars), j)):
        line = ''
        for k in range(1, j + 1):
            line += possible_chars[(int)((i / pow(len(possible_chars), k - 1) % len(possible_chars)))]
        print(line)
```
